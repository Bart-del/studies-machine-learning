Część A (2 punkty)

Wygeneruj wykres funkcji y=f(x)=(a−4)x^2+(b−5)x+(c−6), gdzie a,b,c to trzy ostatnie cyfry Twojego numeru indeksu. Pamiętaj o opisaniu osi wykresu (x oraz y).

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Wykres funkcji z zadania A')

x = np.arange(0.0, 10.0, 0.01)

y = (4-4)*x**2+(6-5)*x +(6-6)

ax.plot(x, y, color='blue', lw=2)

plt.show()
```

Część B (1 punkt)

Na wykresie z zadania 2.2 przedstaw dodatkowo wykres funkcji y=g(x)=exex+1. Linie przestawiające funkcje y=f(x) i y=g(x) powinny znaleźć się na tym samym wykresie, ale powinny różnić się kolorami.

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Wykres funkcji z zadania B')

x = np.arange(0.0, 10.0, 0.01)

y = (4-4)*x**2+(6-5)*x +(6-6)
y1 = (np.e**x)/(np.e**x + 1)

ax.plot(x, y, color='blue', lw=2)
ax.plot(x, y1, color='green', lw=2)

plt.show()
```
